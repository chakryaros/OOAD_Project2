```python
import random

'''
    Create Animal class and has method wakeUp, makeNoise,
    eat, roam, getName
'''
class Animals:
    def __init__(self, Name, type_):
        self.__name = Name
        self.__type = type_
    
    def sleep(self):
#         self.__name = name
        num_random = random.randint(0,3)
        if num_random ==0:
            print("{} falls asleep.".format(self.__name))
        elif num_random ==1:
            print("{} refuses to go to sleep and plays in the dark.".format(self.__name))
        elif num_random ==2:
            print("{} lies down but hasn't fall asleep yet.".format(self.__name))
    
    def wakeUp(self):
        return "..."
    
    def makeNoise(self):
        return "..."
    
    def eat(self):
        return "..."
    
    def roam(self,name):
        return "..."
    
    def getName(self):
        return "..."
    
    
class Canine(Animals):
    
    def __init__(self, name, type_):
        self.__name = name
        super().__init__(name, type_)
    
    def eat(self):
        print("{} eats the steak in a hurry!".format(self.__name))
    
    
    def roam(self,name):
        num_random = random.randint(0,3)
        if num_random ==0:
            print("{} roams his/her area.".format(name))
        elif num_random ==1:
            print("{} is lazy and refuses to roam.".format(name))
        elif num_random ==2:
            print("{} roams a little and stopped.".format(name))


class Dog(Canine):
    
    def __init__(self, name, type_):
        self.__name = name
        self.__type = type_
        super().__init__(name, type_)
        
    def displayType(self):
        print("{} is {}".format(self.__name, self.__type))
    
    def getName(self):
        return self.__name
    
    def makeNoise(self):
        num_random = random.randint(0,3)
        if num_random== 0:
            print("Bark Bark!")
        elif num_random == 1 :
            print("Barkkkkkkk! Because " + self.__name + " is very hungry!")
        else:
            print("Whinn... Because " + self.__name + " is sick!")
        
class Feline(Animals):
    
    def __init__(self, name, type_):
        self.__name = name
        super().__init__(name, type_)
        
    def roam(self,name):
        num_random = random.randint(0,3)
        if num_random ==0:
            print("{} roams his/her area.".format(name))
        elif num_random ==1:
            print("{} is lazy and refuses to roam.".format(name))
        elif num_random ==2:
            print("{} roams a little and stopped.".format(name))
    
    def eat(self):
        print("{} eats the fish in a hurry!".format(self.__name))
        
class Cat(Feline):
    
    def __init__(self, name, type_):
        self.__name = name
        self.__type = type_
        super().__init__(name, type_)
    
    
    def displayType(self):
        print("{} is {}".format(self.__name, self.__type))
        
    def getName(self):
        return self.__name
    
    def makeNoise(self):
        print("Meow Meow!")
    
    def wakeUp(self):
        num_random = random.randint(0,3)
        if num_random == 0:
            print("{} slowly opens his/her eyes and stretches.".format(self.__name))
        elif num_random == 1:
            print("{} wakes up angry because she didn't get enough sleep.".format(self.__name))
        elif num_random == 2:
            print("{} wakes up crying because she had a nightmare.".format(self.__name))
                  
class Pachyderm(Animals):
    
    def __init__(self, name, type_):
        self.__name = name
        super().__init__(name, type_)
        
    def roam(self,name):
        num_random = random.randint(0,3)
        if num_random ==0:
            print("{} roams his/her area.".format(name))
        elif num_random ==1:
            print("{} is lazy and refuses to roam.".format(name))
        elif num_random ==2:
            print("{} roams a little and stopped.".format(name))
    
    def eat(self):
        print("{} eats the banana in a hurry!".format(self.__name))

class Elephant(Pachyderm):
    def __init__(self, name, type_):
        self.__name = name
        self.__type = type_
        super().__init__(name, type_)
    def displayType(self):
        print("{} is {}".format(self.__name, self.__type))
        
    def getName(self):
        return self.__name
    
    def makeNoise(self):
        print("Heah Heah!")
        
    def wakeUp(self):
        print("{} slowly opens his/her eyes and stretches.".format(self.__name))
    
```

```python
#test
if __name__ == '__main__':
    d = Dog("Denny", 'Dog')
    c = Cat("Cassie", "Cat")
    e = Elephant("Elly", "Elephant")
    
    animalList = []
    animalList.append(d)
    animalList.append(c)
    animalList.append(e)
    for i in animalList:
        print()
        i.displayType()
        print()
        i.eat()
        print()
        i.makeNoise()
        print()
        i.eat()
        print()
        i.roam(i.getName())
```
