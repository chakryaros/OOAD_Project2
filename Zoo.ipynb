```python
import random

'''
    Create Animal class and has method wakeUp, makeNoise,
    eat, roam, getName
'''
class Animals:
    def __init__(self, Name, type_):
        self.__name = Name
        self.__type = type_
    
    def sleep(self):
#         self.__name = name
        num_random = random.randint(0,3)
        if num_random ==0:
            print("{} falls asleep.".format(self.__name))
        elif num_random ==1:
            print("{} refuses to go to sleep and plays in the dark.".format(self.__name))
        else:
            print("{} lies down but hasn't fall asleep yet.".format(self.__name))
    
    def wakeUp(self):
        return "..."
    
    def makeNoise(self):
        return "..."
    
    def eat(self):
        return "..."
    
    def roam(self):
        return "..."
    
    def getName(self):
        return "..."
    
    
class Canine(Animals):
    
    def __init__(self, name, type_):
        self.__name = name
        super().__init__(name, type_)
    
    def eat(self):
        print("{} eats the steak in a hurry!".format(self.__name))
    
    
    def roam(self):
        num_random = random.randint(0,3)
        if num_random ==0:
            print("{} roams his/her area.".format(self.__name))
        elif num_random ==1:
            print("{} is lazy and refuses to roam.".format(self.__name))
        else:
            print("{} roams a little and stopped.".format(self.__name))


class Dog(Canine):
    
    def __init__(self, name, type_):
        self.__name = name
        self.__type = type_
        super().__init__(name, type_)
        
    def displayType(self):
        print("{} is {}".format(self.__name, self.__type))
    
    def getName(self):
        return self.__name
    
    def wakeUp(self):
        num_random = random.randint(0,3)
        if num_random == 0:
            print("{} slowly opens his/her eyes and stretches.".format(self.__name))
        elif num_random == 1:
            print("{} wake up with smiley face because he/she slept well.".format(self.__name))
        else:
            print("{} still fall asleep".format(self.__name))
    def makeNoise(self):
        num_random = random.randint(0,3)
        if num_random== 0:
            print("Bark Bark!")
        elif num_random == 1 :
            print("Barkkkkkkk! Because " + self.__name + " is very hungry!")
        else:
            print("Whinn... Because " + self.__name + " is sick!")
        
class Feline(Animals):
    
    def __init__(self, name, type_):
        self.__name = name
        super().__init__(name, type_)
        
    def roam(self):
        num_random = random.randint(0,3)
        if num_random ==0:
            print("{} roams his/her area.".format(self.__name))
        elif num_random ==1:
            print("{} is lazy and refuses to roam.".format(self.__name))
        else:
            print("{} roams a little and stopped.".format(self.__name))
    
    def eat(self):
        print("{} eats the fish in a hurry!".format(self.__name))
        
class Cat(Feline):
    
    def __init__(self, name, type_):
        self.__name = name
        self.__type = type_
        super().__init__(name, type_)
    
    
    def displayType(self):
        print("{} is {}".format(self.__name, self.__type))
        
    def getName(self):
        return self.__name
    
    def makeNoise(self):
        print("Meow Meow!")
    
    def wakeUp(self):
        num_random = random.randint(0,3)
        if num_random == 0:
            print("{} slowly opens his/her eyes and stretches.".format(self.__name))
        elif num_random == 1:
            print("{} wakes up angry because she didn't get enough sleep.".format(self.__name))
        else:
            print("{} wakes up crying because she had a nightmare.".format(self.__name))
            
class Lion(Feline):
    def __init__(self, name, type_):
        self.__name = name
        self.__type = type_
        super().__init__(name, type_)
    
    
    def displayType(self):
        print("{} is a{}".format(self.__name, self.__type))
        
    def getName(self):
        return self.__name
    
    def makeNoise(self):
        print("Roar!")
    
    def wakeUp(self):
        num_random = random.randint(0,3)
        if num_random == 0:
            print("{} is startled by the wakeup.".format(self.__name))
        elif num_random == 1:
            print("{} clearly doesn't want to get up, but does anyway.".format(self.__name))
        else:
            print("{} was already awake and ready for the day!".format(self.__name))
                  
class Pachyderm (Animals):
    
    def __init__(self, name, type_):
        self.__name = name
        super().__init__(name, type_)
        
    def roam(self):
        num_random = random.randint(0,3)
        if num_random ==0:
            print("{} roams his/her area.".format(self.__name))
        elif num_random ==1:
            print("{} is lazy and refuses to roam.".format(self.__name))
        else:
            print("{} roams a little and stopped.".format(self.__name))
    
    def eat(self):
        print("{} eats the banana in a hurry!".format(self.__name))
        
        
class Wolf(Canine):
    def __init__(self, name, type_):
        self.__name = name
        self.__type = type_
        super().__init__(name, type_)
    def displayType(self):
        print("{} is a {}".format(self.__name, self.__type))
        
    def getName(self):
        return self.__name
    
    def makeNoise(self):
        print("AUUHHWOOOOO")
    def wakeUp(self):
        print("{} eagerly opens his/her eyes".format(self.__name))
                    

class Elephant(Pachyderm):
    def __init__(self, name, type_):
        self.__name = name
        self.__type = type_
        super().__init__(name, type_)
    def displayType(self):
        print("{} is {}".format(self.__name, self.__type))
        
    def getName(self):
        return self.__name
    
    def makeNoise(self):
        print("Heah Heah!")
        
    def wakeUp(self):
        num_random = random.randint(0,3)
        if num_random == 0:
            print("{} is startled by the wakeup, and charges the Zookeeper!".format(self.__name))
        elif num_random == 1:
            print("{} clearly doesn't want to get up, but does anyway.".format(self.__name))
        else:
            print("{} was already awake and ready for the day!".format(self.__name))
        
        
class Hippo(Pachyderm):
    def __init__(self, name, type_):
        self.__name = name
        self.__type = type_
        super().__init__(name, type_)
    def displayType(self):
        print("{} is a {}".format(self.__name, self.__type))
        
    def getName(self):
        return self.__name
    
    def makeNoise(self):
        print("MRRRRRREE!")
        
    def wakeUp(self):
        print("{} slowly opens his/her eyes and stretches.".format(self.__name))
        
        
class Zookeepers:
    
    def __init__(self, animalList):
        self.__animalList = animalList
        print("*******Time to Open the Zoo!*******\n")
        
    def wakeAnimals(self):
        print("***Zookeeper starts to wake up all the animals.***")
        
        for animal in self.__animalList:
            print()
            animal.wakeUp()
          
    def callAnimals(self):
        print()
        print("***Time to Call Animals***")
        for animal in self.__animalList:
            print()
            animal.makeNoise()
           
    def feedAnimals(self):
        print()
        print("***Time to Feed Animals***")
        for animal in self.__animalList:
            print()
            animal.eat()
    def exerciseAnimals(self):
        print()
        print("***Time for Animals to Roam***")
        for animal in self.__animalList:
            print()
            animal.roam()
    def shutDownZoo(self):
        print()
        print("***Time for Closing! What a happy day! Zookeeper closes the doors and turns off the lights.***")
        for animal in self.__animalList:
            print()
            animal.sleep()
        
        
        
        
        
        
```
